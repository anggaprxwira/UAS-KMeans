## Pembentukan Model K-Means

```python
x = data.iloc[:, [10,11]].values
x[0:10]
```

Menentukan data x yaitu terpadat pada kolom 10 dan 11 yang merupakan data Placement dan Kill, serta menentukan jumlah array yang digunakan yaitu 10

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

Membuat data x_scaled berdasarkan dari scaler pada data x

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state = 0)
kmeans.fit(x_scaled)
```

Membagi cluster menjadi 10 bagian, dan membuat variabel x_scaled untuk mencari akurasi cluster

```python
y_kmeans = kmeans.fit_predict(x)
y_kmeans
```

Membuat variable y_kmeans berdasarkan dari variable x

```python
data['cluster'] = y_kmeans
```

Mendeskripsikan bahwa data cluster merupakan data y_kmeans

## Pembentukan Visual/Grafik Cluster

## Perhitungan Dataset Menggunakan Knn & Prediksi

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y_kmeans, test_size=0.3, random_state=0)
```

Membagi dataset menjadi 2 jenis data, yaitu data training dan data testing

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

Berfungsi untuk mengambil library standardscaler pada skcith learn preprocessing
dan StandardScaler ini diubah menjadi sc (Pemanggil)
dan 2 data dibawahnya berguna untuk mengatur scale pada x train dan juga x test

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 1)
```

Berfungsi untuk mengimport class KNeighborsClassifier pada sklearn.neighbors dan mengatur KNeighborsClassifier merupakan knn, dan data yang mendekatinya ada 1

```python
knn.fit(x_train, y_train)
```

Digunakan untuk melatih model K-Nearest Neighbors (KNN) dengan data latih dan label yang sesuai.

```python
y_pred = knn.predict(x_test)
y_pred
```

Digunakan untuk mencari nilai y_pred yang merupakan data prekdisi dari x test

```python
from sklearn.metrics import accuracy_score
print ("Akurasi Dataset yang didapatakan : ", accuracy_score(y_test,y_pred)*100)
```

Hasil akurasi dari dataset
